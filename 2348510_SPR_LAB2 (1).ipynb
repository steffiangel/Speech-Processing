(1) Question 1

(a) Generate a basic sinusoidal signal in the time domain.
(For example, you may generate a sine wave with a frequency of 5Hz,
sampled at 1000Hz.)

(b) Plot the time-domain waveform of the signal.

(c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the con-
tinuous frequency spectrum.

(d) Compute the Discrete Fourier Transform (DFT) and plot the discrete
frequency spectrum.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
frequency = 7
sampling_rate = 1000
duration = 1
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Sine wave
amplitude = 1
signal = amplitude * np.sin(2 * np.pi * frequency * t)

plt.figure(figsize=(10, 4))
plt.plot(t, signal)
plt.title('5Hz Sinusoidal Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
signal_fft = np.fft.fft(signal)
fft_magnitude = np.abs(signal_fft)
fft_magnitude = fft_magnitude[:len(fft_magnitude)//2]
freq_axis = np.fft.fftfreq(len(signal), 1/sampling_rate)
freq_axis = freq_axis[:len(freq_axis)//2]

# Plotting the continuous frequency spectrum
plt.figure(figsize=(10, 4))
plt.plot(freq_axis, fft_magnitude)
plt.title('Continuous Frequency Spectrum (Magnitude)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

The DTFT gives a continuous spectrum, while the DFT discretizes it. Peaks are clearly visible in both for simple sinusoidal signals.


(2) Question 2

(a) Generate a composite signal by adding two or more sinusoidal signals of
different frequencies and amplitudes.

(b) Plot the time-domain waveform of the composite signal.

(c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the con-
tinuous frequency spectrum.

(d) Compute the Discrete Fourier Transform (DFT) and plot the discrete
frequency spectrum.

```python
fs = 1000
t = np.linspace(0, 1, fs)

frequency1 = 10
amplitude1 = 1
signal1 = amplitude1 * np.sin(2 * np.pi * frequency1 * t)

frequency2 = 20
amplitude2 = 0.5
signal2 = amplitude2 * np.sin(2 * np.pi * frequency2 * t)

frequency3 = 30
amplitude3 = 0.3
signal3 = amplitude3 * np.sin(2 * np.pi * frequency3 * t)


composite_signal = signal1 + signal2 + signal3
```

```python

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(t, composite_signal, label='Composite Signal', color='blue')
plt.plot(t, signal1, label='Signal 1 (10 Hz)', linestyle='--', alpha=0.7)
plt.plot(t, signal2, label='Signal 2 (20 Hz)', linestyle='--', alpha=0.7)
plt.plot(t, signal3, label='Signal 3 (30 Hz)', linestyle='--', alpha=0.7)
plt.title('Composite Signal Generated from Sinusoidal Signals')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.xlim(0, 1)
plt.ylim(-2, 2)
plt.show()
```

```python
N = len(composite_signal)
frequency = np.fft.fftfreq(N, d=1/fs)
dtft = np.fft.fft(composite_signal)


plt.figure(figsize=(10, 5))
plt.plot(frequency, np.abs(dtft), color='orange')
plt.title('Continuous Frequency Spectrum of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.ylim(0, max(np.abs(dtft)) * 1.1)

plt.show()
```

####Inference:
The presence of sharp peaks at specific frequencies indicates that the composite signal is composed of several discrete frequency components. The taller peaks indicate more significant contributions from those frequencies to the overall signal.The absence of a continuous line or band in the spectrum suggests that the signal does not have a continuous distribution of frequencies. Instead, it is composed of distinct frequency components.

####DTFT VS DFT

DTFT: The Discrete-Time Fourier Transform (DTFT) is used to analyze the frequency content of discrete-time signals. It produces a continuous frequency spectrum, revealing the distribution of energy across all frequencies.

DFT: The Discrete Fourier Transform (DFT) calculates the frequency spectrum of a finite-length sequence of samples. It produces a discrete frequency spectrum with a finite number of frequency bins.DTFT offers finer frequency resolution, while DFT's resolution depends on the sampling rate and the number of samples.



```python


N = len(composite_signal)
dft = np.fft.fft(composite_signal)
frequency_bins = np.fft.fftfreq(N, d=1/fs)


plt.figure(figsize=(10, 5))
plt.stem(frequency_bins, np.abs(dft), basefmt=" ", use_line_collection=True, linefmt='orange', markerfmt='o')
plt.title('Discrete Frequency Spectrum of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.ylim(0, max(np.abs(dft)) * 1.1)
plt.show()
```

(3) Question 3

(a) Generate an exponentially decaying signal.

(b) Plot the time-domain waveform.

(c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the con-
tinuous frequency spectrum.

(d) Compute the Discrete Fourier Transform (DFT) and plot the discrete
frequency spectrum.

(e) Analyze the relationship between the time-domain waveform and the
frequency-domain representation.

```python

fs = 1000
t = np.linspace(0, 5, fs * 5)

decay_rate = 0.95
amplitude = 1.0


exponential_decay_signal = amplitude * np.exp(-decay_rate * t)


plt.figure(figsize=(10, 5))
plt.plot(t, exponential_decay_signal, color='blue')
plt.title('Exponentially Decaying Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(0, 5)
plt.ylim(0, 1.1)
plt.show()
```

```python

N = len(exponential_decay_signal)
frequency = np.fft.fftfreq(N, d=1/fs)
dtft = np.fft.fft(exponential_decay_signal)

plt.figure(figsize=(10, 5))
plt.plot(frequency, np.abs(dtft), color='orange')
plt.title('Continuous Frequency Spectrum of Exponentially Decaying Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.ylim(0, max(np.abs(dtft)) * 1.1)

plt.show()
```

```python

N = len(exponential_decay_signal)
dft = np.fft.fft(exponential_decay_signal)
frequency_bins = np.fft.fftfreq(N, d=1/fs)


plt.figure(figsize=(10, 5))
plt.stem(frequency_bins, np.abs(dft), basefmt=" ", use_line_collection=True, linefmt='orange', markerfmt='o')
plt.title('Discrete Frequency Spectrum of Exponentially Decaying Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.ylim(0, max(np.abs(dft)) * 1.1)
plt.show()
```

The DTFT and DFT of an exponential decay produce a smooth spectrum with energy spread across a range of frequencies.


(4) Question 4

(a) Generate an exponentially decaying signal.

(b) Plot the time-domain waveform.

(c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the con-
tinuous frequency spectrum.

(d) Compute the Discrete Fourier Transform (DFT) and plot the discrete
frequency spectrum.

(e) Analyze the relationship between the time-domain waveform and the
frequency-domain representation.




```python

fs = 1000
t = np.linspace(0, 1, fs, endpoint=False)
frequency = 5
duty_cycle = 0.5


rectangular_wave = np.where(np.mod(t * frequency, 1) < duty_cycle, 1, 0)

plt.figure(figsize=(10, 5))
plt.plot(t, rectangular_wave, color='green')
plt.title('Rectangular Wave Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(0, 1)
plt.ylim(-0.5, 1.5)
plt.show()
```

```python
N = len(rectangular_wave)
dtft = np.fft.fft(rectangular_wave)
frequency_bins = np.fft.fftfreq(N, d=1/fs)
plt.figure(figsize=(10, 5))
plt.plot(frequency_bins, np.abs(dtft), color='orange')
plt.title('Continuous Frequency Spectrum of Rectangular Wave Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.ylim(0, max(np.abs(dtft)) * 1.1)
plt.show()
```

```python
N = len(rectangular_wave)
dft = np.fft.fft(rectangular_wave)
frequency_bins = np.fft.fftfreq(N, d=1/fs)

# Plotting the discrete frequency spectrum
plt.figure(figsize=(10, 5))
plt.stem(frequency_bins[:N // 2], np.abs(dft[:N // 2]), basefmt=" ", use_line_collection=True)
plt.title('Discrete Frequency Spectrum of Rectangular Wave Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.ylim(0, max(np.abs(dft)) * 1.1)
plt.show()
```

For a rectangular wave show harmonics at the fundamental frequency and integer multiples of it. The DFT captures discrete harmonics, while DTFT shows them in continuous frequency space. The shape and strength of the harmonics depend on the signal's duty cycle.

####Inference
Sinusoidal Signal:  DTFT and DFT show sharp peaks at the signal's frequency.

Composite Signal:  DTFT and DFT show multiple sharp peaks for each sinusoidal component.

Exponential Signal:  DTFT and DFT show a smooth, spread-out frequency spectrum without distinct peaks.

Rectangular Wave:  DTFT and DFT display harmonic content at the fundamental frequency and its multiples, depending on the duty cycle.
