# **Linear Time Normalization**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
s1= np.array([2, 3, 4, 6, 8, 7, 6, 5, 4, 3, 2])
s2=np.array([2, 4, 6, 7, 7, 6, 5, 5, 4, 3, 2, 2, 1])
t1=np.linspace(0,1,len(s1))
t2=np.linspace(0,1,len(s2))
```

```python
interp_func = interp1d(t2, s2, kind='linear')  # Linear interpolation
normalized_signal2 = interp_func(t1)
```

```python
plt.figure(figsize=(12, 6))

# Original signals
plt.subplot(2, 1, 1)
plt.plot(t1, s1, label="Signal 1 (Reference)", marker='o')
plt.plot(t2, s2, label="Signal 2 (Test)", marker='x', linestyle='dashed')
plt.title("Original Speech Signals")
plt.xlabel("Normalized Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

# Normalized signals
plt.subplot(2, 1, 2)
plt.plot(t1, s1, label="Signal 1 (Reference)", marker='o')
plt.plot(t1, normalized_signal2, label="Normalized Signal 2", marker='x', linestyle='dashed')
plt.title("Aligned Speech Signals Using Linear Time Normalization")
plt.xlabel("Normalized Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# Compute Alignment Path
alignment_path = list(zip(range(len(s1)), np.round(np.linspace(0, len(s1) - 1, len(s1))).astype(int)))

# Plot Alignment Path
plt.figure(figsize=(6, 6))
x_align, y_align = zip(*alignment_path)
plt.plot(x_align, y_align, marker='o', linestyle='-')
plt.title("Alignment Path")
plt.xlabel("Index in Signal 1")
plt.ylabel("Index in Signal 2")
plt.grid()
plt.show()

# Print Alignment Path
print("Alignment Path:", alignment_path)
```
